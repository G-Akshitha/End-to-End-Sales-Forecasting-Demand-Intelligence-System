## Task 2 - Time Series Analysis & Decomposition

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Order Date'], monthly_sales['Sales'])
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

- The above represents an additive model because the magnitude of the seasonal fluctuations is constant even as the overall trend changes. Therefore the sales series is represented as the sum of trend, seasonal and residual components.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

monthly_ts = monthly_sales.set_index('Order Date')['Sales']

decomposition = seasonal_decompose(monthly_ts, model='additive', period=12)
decomposition.plot()
plt.show()
```

```python
residuals = decomposition.resid.dropna()

print("Highest positive residual:")
print(residuals.idxmax(), residuals.max())

print("Highest negative residual:")
print(residuals.idxmin(), residuals.min())
```

1. **Trend:** Sales show an overall upward trend, indicating growth over the four-year period
2. **Seasonality:** A recurring seasonal pattern is present implying  that certain months consistently influence sales.
3. **Residuals:** Most residuals fluctuate around zero indicating that the trend and seasonal components capture a substantial portion of the sales pattern. May 2017 had the largest positive residual (20804), while September 2017 had the largest negative residual (-24837)

```python
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(monthly_ts)
print("ADF Statistic:", adf_result[0])
print("p-value: ", adf_result[1])
for key, value in adf_result[4].items():
    print(f"Critical Value ({key}): {value}")
```

p-value (0.696) is greater than the significance level 0.05 which fails to reject the Null Hypothesis. This indicates that the data is non-stationary. The ADF Statistic is also greater than 5% critical value which also indicates we fail to reject the Null Hypothesis and that the data is non-stationary.

Stationarity: A time series is stationary when its statistical behaviour remains relatively stable over time i.e its average level and variability do not systematically change over time.

```python
# First-order differencing
monthly_diff = monthly_ts.diff().dropna()

# ADF test on differenced data
result_diff = adfuller(monthly_diff)

print("ADF Statistic:", result_diff[0])
print("p-value:", result_diff[1])
for key, value in result_diff[4].items():
    print(f"Critical Value ({key}): {value}")
```

- p-value is less than 0.05 which means we reject the null hypothesis and the differenced series is stationary. The critical value comparision (-3.977 < -2.946) confirms that we reject the null hypothesis. This confirms the data is stationary.

### SARIMA (Statistical Model)

```python
import itertools
from statsmodels.tsa.statespace.sarimax import SARIMAX

p = q = range(0,3)
d = [1]

# Seasonal parameters
P = Q = range(0,2)
D = [0, 1]
m = 12

pdq = list(itertools.product(p,d,q))
seasonal_pdq = list(itertools.product(P,D,Q,[m]))
results = []

for order in pdq:
    for seasonal_order in seasonal_pdq:
        try:
            model = SARIMAX(monthly_ts, 
                            order=order, 
                            seasonal_order=seasonal_order, 
                            enforce_stationarity=False,
                            enforce_inertibility=False)
            fitted_model = model.fit(disp=False)
            results.append([order, seasonal_order, fitted_model.aic])
        except:
            continue
```

```python
import pandas as pd

results_df = pd.DataFrame(results,columns=['order', 'seasonal_order', 'AIC'])
results_df.sort_values('AIC').head(10)
```

SARIMA(1,1,2)(0,1,1,12) was selected because it had the lowest AIC value of 447.24 among the tested candidate models.

```python
final_model = SARIMAX(
    monthly_ts,
    order=(1, 1, 2),
    seasonal_order=(0, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_result = final_model.fit(disp=False)

print(sarima_result.summary())
```

```python
forecast = sarima_result.get_forecast(steps=3)

forecast_mean = forecast.predicted_mean
confidence_interval = forecast.conf_int(alpha=0.05)

print("3-Month Forecast:")
print(forecast_mean)

print("\n95% Confidence Intervals:")
print(confidence_interval)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Actual monthly sales
plt.plot(
    monthly_ts.index,
    monthly_ts.values,
    label='Actual Sales'
)

# Forecasted sales
plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    marker='o',
    label='Forecasted Sales'
)

# 95% confidence interval
plt.fill_between(
    confidence_interval.index,
    confidence_interval.iloc[:, 0],
    confidence_interval.iloc[:, 1],
    alpha=0.3,
    label='95% Confidence Interval'
)

plt.title('Actual vs 3-Month SARIMA Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

## Task 4 - Product Category & Region Level forecasting

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor

def xgboost_segment_forecast(segment_data, steps=3):

    # Aggregate order-level sales into monthly sales
    monthly = (
        segment_data
        .set_index('Order Date')['Sales']
        .resample('M')
        .sum()
        .reset_index()
    )

    # Create lag features
    monthly['Lag_1'] = monthly['Sales'].shift(1)
    monthly['Lag_2'] = monthly['Sales'].shift(2)
    monthly['Lag_3'] = monthly['Sales'].shift(3)

    # Rolling mean using only previous months
    monthly['Rolling_Mean_3'] = (
        monthly['Sales']
        .shift(1)
        .rolling(3)
        .mean()
    )

    # Calendar features
    monthly['Month'] = monthly['Order Date'].dt.month
    monthly['Quarter'] = monthly['Order Date'].dt.quarter
    monthly['Season'] = monthly['Month'].apply(get_season)

    # One-hot encode season
    monthly = pd.get_dummies(
        monthly,
        columns=['Season'],
        dtype=int
    )

    # Remove rows containing NaN from lag creation
    model_data = monthly.dropna().copy()

    X = model_data.drop(columns=['Order Date', 'Sales'])
    y = model_data['Sales']

    # Train XGBoost
    model = XGBRegressor(
        n_estimators=100,
        learning_rate=0.05,
        max_depth=3,
        random_state=42
    )

    model.fit(X, y)

    # Store historical sales for recursive forecasting
    sales_history = monthly['Sales'].tolist()

    # Create next 3 monthly dates
    future_dates = pd.date_range(
        start=monthly['Order Date'].max() + pd.offsets.MonthEnd(1),
        periods=steps,
        freq='M'
    )

    predictions = []

    for date in future_dates:

        month = date.month
        quarter = date.quarter
        season = get_season(month)

        future_row = pd.DataFrame({
            'Lag_1': [sales_history[-1]],
            'Lag_2': [sales_history[-2]],
            'Lag_3': [sales_history[-3]],
            'Rolling_Mean_3': [
                np.mean(sales_history[-3:])
            ],
            'Month': [month],
            'Quarter': [quarter],
            'Season_Fall': [1 if season == 'Fall' else 0],
            'Season_Spring': [1 if season == 'Spring' else 0],
            'Season_Summer': [1 if season == 'Summer' else 0],
            'Season_Winter': [1 if season == 'Winter' else 0]
        })

        # Match training columns exactly
        future_row = future_row.reindex(
            columns=X.columns,
            fill_value=0
        )

        prediction = model.predict(future_row)[0]

        predictions.append(prediction)

        # Add prediction for forecasting the next month
        sales_history.append(prediction)

    return pd.DataFrame({
        'Date': future_dates,
        'Forecast': predictions
    })
```

### Z-Score based anomaly detection

```python
window = 12

weekly_sales['Rolling_Mean'] = (
    weekly_sales['Sales']
    .rolling(window=window, min_periods=window)
    .mean()
)

weekly_sales['Rolling_Std'] = (
    weekly_sales['Sales']
    .rolling(window=window, min_periods=window)
    .std()
)

weekly_sales['Z_Score'] = (
    (weekly_sales['Sales'] - weekly_sales['Rolling_Mean'])/ weekly_sales['Rolling_Std']
)

weekly_sales['Z_Anomaly'] = (weekly_sales['Z_Score'].abs() > 2)
```

```python
z_anomalies = weekly_sales[weekly_sales['Z_Anomaly']]

print(z_anomalies[['Order Date', 'Sales', 'Z_Score']])
```

## Task 6 — Product Demand Segmentation using Clustering

### Total sales volume and average order value

```python
subcat_features = (
    df.groupby('Sub-Category')
      .agg(
          Total_Sales=('Sales', 'sum'),
          Average_Order_Value=('Sales', 'mean')
      )
      .reset_index()
)

subcat_features.head()
```

### Calculate sales volatality

```python
monthly_subcat = (
    df.groupby([
        'Sub-Category',
        pd.Grouper(key='Order Date', freq='M')
    ])['Sales']
    .sum()
    .reset_index()
)
```

```python
volatility = (
    monthly_subcat
    .groupby('Sub-Category')['Sales']
    .std()
    .reset_index(name='Sales_Volatility')
)

subcat_features = subcat_features.merge(
    volatility,
    on='Sub-Category'
)
```

### Year-Over-Year sales growth

```python
yearly_subcat = (
    df.groupby([
        'Sub-Category',
        df['Order Date'].dt.year.rename('Year')
    ])['Sales']
    .sum()
    .reset_index()
)
```

```python
yearly_subcat['YoY_Growth'] = (
    yearly_subcat
    .groupby('Sub-Category')['Sales']
    .pct_change() * 100
)
```

```python
avg_growth = (
    yearly_subcat
    .groupby('Sub-Category')['YoY_Growth']
    .mean()
    .reset_index(name='Sales_Growth_Rate')
)

subcat_features = subcat_features.merge(
    avg_growth,
    on='Sub-Category'
)
```

```python
subcat_features
```

### Scaling the features

```python
from sklearn.preprocessing import StandardScaler
features = ['Total_Sales','Sales_Growth_Rate','Sales_Volatility','Average_Order_Value']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(subcat_features[features])
```

### Elbow Method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

K = range(1, 10)

for k in K:
    kmeans = KMeans(n_clusters=k,random_state=42,n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, marker='o')

plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

### Applying K-Means

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

subcat_features['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_profile = (
    subcat_features
    .groupby('Cluster')[features]
    .mean()
    .round(2)
)

cluster_profile
```

```python
cluster_labels = {
    0: 'High Volume, Stable Demand',
    1: 'Low Volume, Stable Demand',
    2: 'Rapidly Growing Demand',
    3: 'High-Value, Volatile Demand'
}

subcat_features['Cluster_Label'] = (
    subcat_features['Cluster'].map(cluster_labels)
)

subcat_features[
    ['Sub-Category', 'Cluster', 'Cluster_Label']
]
```

### Applying PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

subcat_features['PC1'] = X_pca[:, 0]
subcat_features['PC2'] = X_pca[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

for cluster in sorted(subcat_features['Cluster'].unique()):

    cluster_data = subcat_features[
        subcat_features['Cluster'] == cluster
    ]

    plt.scatter(
        cluster_data['PC1'],
        cluster_data['PC2'],
        s=100,
        label=cluster_labels[cluster]
    )

    # Show sub-category names
    for _, row in cluster_data.iterrows():
        plt.annotate(
            row['Sub-Category'],
            (row['PC1'], row['PC2']),
            xytext=(5, 5),
            textcoords='offset points',
            fontsize=8
        )

plt.title('Product Demand Segmentation using K-Means and PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

- The PCA visualization shows distinct separation among the four K-Means clusters. Most stable-demand products are grouped relatively close together, while rapidly growing and high-value volatile products appear farther from the main groups, indicating substantially different demand characteristics
